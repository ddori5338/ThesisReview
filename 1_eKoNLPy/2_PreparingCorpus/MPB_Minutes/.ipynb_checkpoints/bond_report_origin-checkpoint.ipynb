```python
import requests
from bs4 import BeautifulSoup
from PyPDF2 import PdfFileReader
import sys
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.converter import XMLConverter, HTMLConverter, TextConverter
from pdfminer.layout import LAParams
import io
from urllib.request import urlopen
```

```python
def content_to_txt(pdf_content, name):
    #file_res = requests.get(pdf_link)
    with open(name,'w', encoding='utf-8') as f:
        f.write(pdf_content)
```

```python
def download_file(download_url,name):
    response = urlopen(download_url)
    file = open("{}.pdf".format(name), 'wb')
    file.write(response.read())
    file.close()
    #print("Download Completed")
```

```python
def pdfparser(filename):

    fp = open(filename, 'rb')
    rsrcmgr = PDFResourceManager()
    retstr = io.StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    # Create a PDF interpreter object.
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    # Process each page contained in the document.

    for page in PDFPage.get_pages(fp):
        interpreter.process_page(page)
        data = retstr.getvalue()
    fp.close()

    return data.replace('\n','')
```

```python
last_page_num = '116'
count_i=1110
#final_list = []
for num in range(38,int(last_page_num)+1):
    page_param = '?&page={}'.format(num)
    url = 'https://finance.naver.com/research/debenture_list.nhn' + page_param
    resp = requests.get(url)
    soup = BeautifulSoup(resp.content)
    
    for i in range(len(soup.find('table',class_='type_1').find_all('tr'))-2):
    #for i in range(3):
        
        if soup.find('table',class_='type_1').find_all('tr')[i+2].find('td',class_='date') == None:
            continue
        
        #new_dict = {}
        
        date = soup.find('table',class_='type_1').find_all('tr')[i+2].find('td',class_='date').text
        corp = soup.find('table',class_='type_1').find_all('tr')[i+2].find_all('td')[1].text
        title = soup.find('table',class_='type_1').find_all('tr')[i+2].find('td').find('a').text
        title = title.replace('?','')
        title = title.replace('!','')
        title = title.replace('(','')
        title = title.replace(')','')
        title = title.replace('{','')
        title = title.replace('}','')
        title = title.replace('[','')
        title = title.replace(']','')
        title = title.replace(':','')
        title = title.replace(';','')
        title = title.replace(',','')
        title = title.replace('*','')
        title = title.replace('/','')
        title = title.replace('+','')
        title = title.replace('=','')
        title = title.replace('"','')
        title = title.replace("'",'')
        title = title.replace("<",'')
        title = title.replace(">",'')
        

        #new_dict['date'] = date
        #new_dict['corp'] = corp
        #new_dict['content'] = content
        #final_list.append(new_dict)
        
        pdf_href = soup.select('div.box_type_m table.type_1')[0].find_all('tr')[i+2].find('td',class_='file').find('a')['href']
        download_file(pdf_href, title)
        
        file_name = title+'.pdf'
        #content = pdfparser(file_name)
        
        try:
            txt_file_name = date+'_'+corp+'_'+title+'.txt'
            content_to_txt(pdfparser(file_name), txt_file_name)
        except AttributeError:
            print('attribute error로 파일 변환 실패')
            pass
        except OSError:
            print('OSError로 파일 변환 실패')
            pass
#         except:
#             print('다른 에러로 파일 변환 실패')
#             pass

        count_i+=1
        print(file_name,', {}번째 파일 변환 완료'.format(count_i))
```
