```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
startPage=54
endPage=517
for i  in range(startPage,endPage): #54~515
    url='https://finance.naver.com/marketindex/interestDailyQuote.nhn?marketindexCd=IRR_CALL&page={}'.format(i)
    resp=requests.get(url)
    soup=BeautifulSoup(resp.content)
    rowList = soup.find('tbody').findAll('tr')
    for row in rowList:
        date = row.find('td', class_='date').text.strip()
        dateList.append(date)
        num = row.find('td', class_='num').text.strip()
        numList.append(num)
    print('{} 정보를 가져오는 중입니다.'.format(date))
    df = pd.DataFrame({'date': dateList, 'close':numList})
    df.to_csv('callRate.csv', index=False, header=False ,mode='a')
```
