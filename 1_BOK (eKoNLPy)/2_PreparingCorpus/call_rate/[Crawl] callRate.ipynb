```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def callRate(startPage, endPage):
    all_info=[]
    for i  in range(startPage,endPage): #54~515
        url='https://finance.naver.com/marketindex/interestDailyQuote.nhn?marketindexCd=IRR_CALL&page={}'.format(i)
        resp=requests.get(url)
        soup=BeautifulSoup(resp.content)
        rowList = soup.find('tbody').findAll('tr')
        
        for row in rowList:
            date = row.find('td', class_='date').text.strip().replace('.','-')
            num = row.find('td', class_='num').text.strip()
            all_info.append([date, num])
            print('{} 정보를 가져오는 중입니다.'.format(date))
            df = pd.DataFrame({'date': [date], 'callRate':[num]})
            
            if all_info.index([date, num])==0:                
                df=df.to_csv('callRate.csv', index=False, columns=['date','callRate'])
            else:
                df=df.to_csv('callRate.csv', index=False, header=False, mode='a')
    return all_info
    print('='*50)
    print('모든 콜금리 정보를 저장했습니다.')
```

```python
all_info = callRate(55, 521) #55~520
```

```python
df = pd.DataFrame(all_info, columns=['date', 'callRate'])
```

```python
def label(df):
    dfCompare = df.shift(-21)
    length = len(dfCompare.dropna())
    
    origin = df['callRate'].astype(float)
    compare = dfCompare['callRate'].astype(float)
    
    label = origin[:length]-compare[:length]

    for idx in range(length):
        if label[idx] > 0:
            label[idx] = 'pos'
        elif label[idx] < 0:
            label[idx] = 'neg'
    dfLabel = pd.concat([df['date'], label], axis=1).dropna()
    dfLabel.columns = ['date','label']
    dfLabel.to_csv('label.csv')
    return dfLabel
```

```python
label = label(df)
```
