```python
import os
```

```python
import pandas as pd
from ekonlpy.sentiment import MPCK

def text2tokens(text):
    return MPCK().tokenize(text)

def text2ngram(text):
    mpck = MPCK()
    bef_tokens = mpck.tokenize(text)
    ngrams = mpck.ngramize(bef_tokens)
    return ngrams
```

```python
text2ngram("10월 정치 불확실성, 미완의 종결")
```

```python
import pandas as pd
path = r'C:\Users\student\Documents\bondreport_txt'
columns = ['date', 'firm','title','report', 'tokens','ngrams']

data = []


for title in os.listdir(path):
    date = title.split("_")[0]
    firm = title.split("_")[1]
    name = title.split("_")[2].replace(".txt","")
    file = path + '\\' + title
    
    try:
        with open(file, 'r', encoding = 'utf8') as f:
            content = f.read()
            tokens = text2tokens(content)
            ngrams = text2ngram(content)
            print("{}{} finished".format(date,name))
#         content = open(file, 'r', encoding = 'utf8')
#         tokens = text2tokens(content.read())
#         ngrams = text2ngram(content.read())
            data.append([date, firm, name, content, tokens, ngrams])
#         
    except:
        print("{} error".format(file))
        
df = pd.DataFrame(data, index = [i for i in range(len(os.listdir(path)))], columns = columns)
```
